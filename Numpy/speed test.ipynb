```python
import numpy as np
import time
size=1000000
l1=list(range(size))
l2=list(range(size))
start=time.time()
add=[x+y for x,y in zip(l1,l2)]
end=time.time()
print(end-start)
```

```python
size=1000000
l1=np.array(list(range(size)))
l2=np.array(list(range(size)))
start=time.time()
add=l1+l2
end=time.time()
print(end-start)
```

```python
l3=list(range(size))
start=time.time()
mul=[x**2 for x in l3]
print(mul[0:10])
end=time.time()
print(end-start)
```

```python
l3=np.array(list(range(size)))
start=time.time()
mul=l3**2
print(mul[0:10])
end=time.time()
print(end-start)
```
