```python
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
labels=["Sachin", "Kohli","Sehwag","Yuvraj"]
colors=['#F52779','#F5E427','#B027F5','#F52779']
explode=[0.1,0.1,0.1,0.1]
runs=[1200,3455,4566,3455]
plt.title("Career runs of Indian batsman")
plt.pie(runs,labels=labels,explode=explode,colors=colors,wedgeprops={'edgecolor':'black','linewidth':2,'linestyle':'--'},shadow=True,startangle=140,  autopct='%1.1f%%')
```
